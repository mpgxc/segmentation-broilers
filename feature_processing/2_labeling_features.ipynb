```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('tmp/1_feature_extraction.csv')
broilers_weights = pd.read_csv('galinhas_pesos.csv')
```

```python
data['target'] = -1
```

```python
data.head()
```

```python
for anilha, peso in zip(broilers_weights.anilhas,  broilers_weights.pesos):

    try: 
        index, *_ = data[data.label == anilha].index
        data.target.iloc[index] = peso
    except:
        continue
```

```python
broilers_weights.anilhas.value_counts()
```

```python
query_filter = data.label.apply(lambda value: value in broilers_weights.anilhas.to_list())
data[query_filter]
```
